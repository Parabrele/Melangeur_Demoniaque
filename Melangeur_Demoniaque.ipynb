```python
import torch
from torchvision import transforms

import images
from perlin_noise import perlin_noise, smooth
import melangeur_demoniaque as md
```

```python
im1 = images.load_image("../../images/Eva_01.jpg", size=720)
im2 = images.load_image("../../images/Eva_02.jpg", size=720)
```

```python
map_1 = md.create_map(im1, 0.2)
map_2 = md.create_image_map(im1, 0.2)
map_3 = md.create_outline_map(im1, 0.2)
map_4 = md.create_smooth_outline_map(im1, 0.2)
map_5 = md.create_perlin_image_map(im1, im2, 0.2, 0.7)
map_6 = md.create_perlin_edge_map(im1, im2, 0.2, 0.7)
map_7 = md.create_image_edge_map(im1, im2, 0.2)
```

```python
images.imshow([map_1, map_2, map_3, map_4, map_5, map_6, map_7])
```

```python
im2 = images.load_image("../../images/drunk_v3.jpg", size=720)
im1 = images.load_image("../../images/bacon_01.jpg", size=720)
```

```python
noise = md.create_perlin_edge_map(im1, im2, 0.2, 0.7)
demon = md.blend(im1, im2, noise)

demon = transforms.Resize((900, 1280))(demon)
demon = demon.permute(0, 1, 3, 2).flip(2)

images.imshow(demon)
images.save_image(demon, 'demonic_03.jpg')
```
